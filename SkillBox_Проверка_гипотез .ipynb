# Проверка выборки на нормальное распределение

Для проверки того, что выборка имеет нормальное распределение, можно использовать различные статистические тесты, например, Шапиро-Уилка или Колмогорова-Смирнова.

Вот пример использования теста Колмогорова-Смирнова:


```python
from scipy.stats import ks_2samp
import numpy as np

# Создаем выборку из нормального распределения
sample = np.random.normal(0, 1, 100)

# Проверяем, является ли выборка нормально распределенной
print(ks_2samp(sample, np.random.standard_normal(size=10000)))

# В этом коде мы создаем выборку из стандартного нормального распределения и
# сравниваем ее с выборкой из 10 000 элементов, генерированной из того же распределения.
# Если обе выборки были взяты из одного и того же распределения,
# то тест должен показать результат близкий к 1.
# В противном случае, если выборки были взяты из разных распределений, результат теста будет меньше 1.
```

Результат выполнения теста Колмогорова-Смирнова на выборке показывает, что нет оснований отвергнуть гипотезу о том, что выборка была взята из стандартного нормального распределения, так как значение pvalue больше критического значения 0.05. Значение statistic показывает расстояние между двумя функциями распределения, в данном случае оно равно 0.0871. Значения statistic_location и statistic_sign не используются в данном контексте и могут быть проигнорированы.

# P-значение (P-value)
Р-значение (P-value) — вероятность ошибки при отклонении нулевой гипотезы (ошибки первого рода).

Вероятность допущения ошибки второго рода = $\beta$.<br>
Мощность критерия = $1 - \beta$.
Мощность — это вероятность отклонения нулевой гипотезы, когда она ложна.

## U-критерий Манна-Уитни

### Ограничения критерия Манна — Уитни
1. Объёмы исследуемых выборок не должны быть больше 60
2. Минимальное число наблюдений — 3 наблюдения в каждой выборке
3. Объёмы выборок не должны сильно различаться

Вариационный ряд - это упорядоченный ряд наблюдений

```python
from scipy.stats import rankdata

# пример массива данных
array = np.array([21, 16, 34, 34, 5])
# Считаем ранги
rankdata(array)
```

```python
sample1 = np.array([7, 6, 4, 5])
sample2 = np.array([2, 8, 3, 6, 4, 7, 5])
```

```python
sample_united = list(sample1)+list(sample2)
sample_united
```

```python
rank_united= rankdata(sample_united)
rank_united
```

```python
rank1 = rank_united[:len(sample1)]
rank2 = rank_united[len(sample1):]
```

```python
func = lambda sample1,sample2: (len(sample1), rank1.sum()) \
       if rank1.sum() >= rank2.sum() \
       else (len(sample2), rank2.sum())
```

```python
n = func(sample1,sample2)
n
```

```python
U = len(sample1)*len(sample2) + n[0]*(n[0]+1)/2  - n[1]
U
```

## Задача 1

Известно, что в генеральной совокупности средняя продолжительность жизни мужчин в Европе составляет $µ_0 = 75,1$ года. Вам дана некоторая выборка мужчин, продолжительность жизни которых составляет 56, 81, 72, 65, 67, 62 и 70 лет соответственно. Проверьте нулевую гипотезу о том, что средняя продолжительность мужчин в выборке не отличается от среднего совокупности: $µ_0 = µ_1$.

```python
# Используем одновыборочный t-критерий Стъюдента

sample1 = np.array([56, 81, 72, 65, 67, 62, 70])
mu_0 = 75.1

t = (mu_0 - sample1.mean()) / (sample1.std()/np.sqrt(len(sample1)))
print(f'выборочное среднее: {sample1.mean()}')
print(f'число степеней свободы = {len(sample1)-1}')
print(f'значение t-критерия = {t}' )
```

Т.к. расчетное значение больше табличного 2.711 > 2.447. Следовательно гипотезу $H_0$ можно отклонить и принять альтернативную гипотезу $H_1$ о том, что в генеральной совокупности средняя продолжительность жизни мужчин в Европе составляет менее 75,1 года.

## Задача 2

По выборке объёма $n = 35$ найден средний вес изделий, изготовленных на заводе: $µ_1 = 144$ г. Известно, что по ГОСТу средний вес деталей составляет $µ_0 = 135$ г. Среднеквадратичное в выборке равняется $σ = 50$г. Проверьте нулевую гипотезу о равенстве средних: $µ_0 = µ_1$.

```python
# Используем одновыборочный t-критерий Стъюдента

mu_0 = 135
sample_mean = 144
n = 35

t = (sample_mean - mu_0) / (sample_mean/np.sqrt(n))
print(f'выборочное среднее: {sample_mean}')
print(f'число степеней свободы = {n-1}')
print(f'значение t-критерия = {t}' )
```

Т.к. расчетное значение меньше табличного 0.37 < 2.032. Следовательно у нас нет оснований, чтобы отклонить гипотезу $H_0$.

### Доверительные интервалы

```python
def normal_confidence_interval(n, mean, std, alpha=0.05):
    '''
    функция вычисляет доверительный интервал для уровня значимости 0.05
    с помощью нормального распределения
    '''
    # число степеней свободы
    df = n - 1

    # стандартная ошибка среднего
    std_err = std / np.sqrt(df)

    z_value = stats.norm.ppf(1 - alpha / 2)
    return [mean - z_value * std_err, mean + z_value * std_err]


def t_student_confidence_interval(n, mean, std, alpha=0.05):
    '''
    функция вычисляет доверительный интервал для уровня значимости 0.05
    с помощью t-распределения
    '''

    # стандартная ошибка среднего
    standard_error = std / np.sqrt(n)

    # Вычислим доверительный интервал с помощью t-распределения
    alpha = 0.05  # уровень доверия 95%
    t_critical = stats.t.ppf(1 - alpha/2, df=n-1)  # критическое значение t для двустороннего доверительного интервала
    margin_of_error = t_critical * standard_error
    return [mean - margin_of_error, mean + margin_of_error]
```

## Задача 5

Дана реализация двух случайных ненормально распределённых величин X и Y. Проверьте нулевую гипотезу о равенстве средних значений: µx = µy. Подсказка: воспользуйтесь непараметрическим тестом (критерием Манна — Уитни).

X:	-2	1	4	-1	5	7	9	0	12 <br>
Y:	6	-3	2	1	0	4	2

```python
sample1 = np.array([ -2, 1, 4, -1, 5, 7, 9, 0, 12])
sample2 = np.array([6, -3, 2, 1, 0, 4, 2])
sample_united = list(sample1)+list(sample2)
rank_united= rankdata(sample_united)
rank1 = rank_united[:len(sample1)]
rank2 = rank_united[len(sample1):]
```

```python
func = lambda sample1,sample2: (len(sample1), rank1.sum()) \
       if rank1.sum() >= rank2.sum() \
       else (len(sample2), rank2.sum())
```

```python
n = func(sample1,sample2)
U = len(sample1)*len(sample2) + n[0]*(n[0]+1)/2  - n[1]
print('объем 1-й выборки N1:',len(sample1))
print('объем 2-й выборки N2:',len(sample2))
print('расчетное значение U-критерия:',U)
```
