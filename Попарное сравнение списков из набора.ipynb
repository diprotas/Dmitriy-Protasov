```python
# сравнение двух списков
def returnMatches(list1):
    united_list = [j for i in list1 for j in i]
    matrix = np.ones((len(united_list), len(united_list)))
    for i in range(len(united_list)):
        for j in range(len(united_list)):
            if united_list[i] != united_list[j]:
                matrix[i][j] = 0
    return matrix
# требуется сравнить попарно списки из набора:
list1 = [['a', 'b', 'c', 'g', 'h'], ['b', 'c', 'd', 'g'], ['c', 'e', 'a', 'f','h']]


returnMatches(list1)
```
