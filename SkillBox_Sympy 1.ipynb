```python
from sympy import *
```

```python
x = Symbol('x')
expr = (x - 6*x/(x+2)) / ((x-4)/(x+2))
```

```python
expr = expr.simplify().subs({x : 5})
expr
```

```python
a, b = symbols('a, b')
```

```python
expr1 = (4/a**2 - 9/b**2) / (a*b)**2
```

```python
expr1 = expr1.simplify().subs({a : 1, b : 2})
expr1
```
