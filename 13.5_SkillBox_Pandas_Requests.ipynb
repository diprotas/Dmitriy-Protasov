#  Requests

```python
import requests
import pandas as pd
```

## GET

```python
url = 'https://itunes.apple.com/search'
params = {'media': 'music', 'term': 'manizha', 'limit': 30}
response = requests.get(url, params=params)
response
```

```python
response.status_code
```

```python
response.text
```

```python
response.json()
```

```python
results = response.json()['results']
results
```

```python
df = pd.DataFrame(results)
df
```

```python
df.primaryGenreName.value_counts()
```

```python
help(df.to_excel)
```

```python
df.to_excel('data/iTunes_api.xlsx', sheet_name='Manizha', index_label='id')
```

```python
help(df.to_csv)
```

```python
df.to_csv('data/iTunes_api.csv', sep='\t', index=False)
```

# Список университетов

С помощью этого API можно получить список университетов в указанной стране.

```python
url = 'http://universities.hipolabs.com/search?'
```

```python
params = {'country': 'canada'}
response = requests.get(url, params=params)
response
```

```python
results = response.json()
results
```

```python
df = pd.DataFrame(results)
df
```

```python
df.info()
```

```python
df.columns
```

```python
df['state-province'].unique()
```

```python
df.groupby(['state-province']).agg('count')
```

Для удобства давайте представим столбец state-province как категориальную переменную (category) и установим порядок, в котором хотим просматривать.

Это не является строго обязательным, но помогает поддерживать желаемый порядок при работе с данными.


```python
df["state-province"] = df["state-province"].astype("category")
```

```python
df.pivot_table(index=['state-province'])
```

```python
# 'country', 'domains', 'alpha_two_code', 'state-province', 'web_pages', 'name'
dict_func = {'country': ['count'],
             'domains': ['count'],
             'state-province' : ['count'],
             'web_pages': ['count']
             }

print(df.agg(dict_func))
df.agg()
```
