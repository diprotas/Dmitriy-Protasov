# Преобразование категориальных данных

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
df = pd.read_csv('/content/drive/MyDrive/SkillBox/Feature_Processing/vehicles_dataset_upd4.csv')
df.head()
```

```python
ohe = OneHotEncoder(sparse_output=False)
ohe
```

```python
df.fuel.unique()
```

```python
# кодируем категориальный признак тип топлива
ohe.fit(df[['fuel']])
```

```python
ohe_fuel = ohe.transform(df[['fuel']])
ohe_fuel
```

```python
ohe_fuel.shape
```

```python
ohe.categories_
```

```python
ohe.inverse_transform(ohe_fuel)
```

```python
#Вывести предлагаемые названия признаков, которые могут пойти в датафрейм
#fuel_ означает принадлежность к признаку, а после нижнего подчеркивания расположены названия категорий
ohe.get_feature_names_out()#get_feature_names()
```

```python
df[ohe.get_feature_names_out()] = ohe_fuel
```
