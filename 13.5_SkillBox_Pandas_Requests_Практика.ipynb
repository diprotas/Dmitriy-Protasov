**Задача 2**

```python
# 1.
url = 'https://itunes.apple.com/search'
params = {'term': 'Einaudi', 'limit': 150, 'media': 'music'}
response = requests.post(url, data=params)
response_json = response.json()
response_json
```

```python
# 2.
df = pd.DataFrame(response_json['results'])
df
```

```python
df.shape()
```

```python
# 3.

df.to_excel('data/Ludovico Einaudi.xlsx', sheet_name='track', index_label='id')
```

```python
# 4.

df.to_csv('data/Ludovico Einaudi.csv', sep='|')
```

# Узнаём температуру в городе с помощью https://api.openweathermap.org

```python
import requests

city = 'Москва'

url = 'https://api.openweathermap.org/data/2.5/weather?q='+city+'&units=metric&lang=ru&appid=79d1ca96933b0328e1c7e3e7a26cb347'

weather_data = requests.get(url).json() #json-структура

temperature = round(weather_data['main']['temp']) #из json-структуры вынимаем значение
temperature_feels = round(weather_data['main']['feels_like'])

print('Сейчас в городе', city, str(temperature), '°C')
print('Ощущается как', str(temperature_feels), '°C')
```
