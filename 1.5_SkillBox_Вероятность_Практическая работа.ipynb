```python
def coinToss(number):
    recordList, proportionList, orli, reshki = [], [], 0, 0
    rangeList = range(1, number+1) #количество бросков
    for i in rangeList:
        flip = random.randint(0, 1) #возвращает либо 0 (орёл), либо 1 (решка)
        if flip == 0:
            recordList.append("О")
            orli += 1
        else:
            recordList.append("Р")
            reshki += 1
        proportionList.append(orli/i)

    print("Число испытаний: "+ str(number))
    print("Доля орлов: " + str(orli/i) + " Доля решек: " + str(reshki/i))
    print((rangeList, proportionList))


    #Построение графика
    plt.figure(figsize=(40, 10), dpi=80)
    plt.plot(rangeList, proportionList)
    # plt.plot(x, vek_polyn(x)) # наш интерполяционный полином
    plt.tick_params(axis='both', which='major', labelsize=10)
    plt.xticks(np.arange(1, number+4, step=5))
    plt.yticks(np.arange(0, 1.1, step=0.1))
```

Создадим функцию, которая считает частоту выпадения орешки в n бросках. Затем векторизуем её, чтобы она могла работать с numpy-векторами.

```python
def numb_res_coin(n):
  '''
  Функция возвращает долю выпадений решки
  при бросании монеты n раз
  '''
  return np.random.randint(0,2,n).sum()/n

# numpy вектоирзация функции
vek_numb_res_coin = np.vectorize(numb_res_coin)
```

Функция `diceLoss` моделирует бросок двух игральных кубиков некоторое количество раз (`number`). В результате выводится вероятность того, что сумма на кубиках будет больше параметра функции `k` (считается доля выпадений, на которых сумма превысит `k`, относительно всех сумм).

1. Поэкспериментируйте с разными `number` и `k`. К каким значениям будет стремиться вероятность при фиксированном `k` и больши́х `number`? Что будет, если брать `k` = 6, 8? Как полученные вероятности будут связаны с теоретическими вероятностями?

2. К чему будет стремиться общая сумма бросков, поделённая на их количество? К чему стремится средняя разность текущего броска и полученного среднего значения? Напишите код на Python на основе исходной функции и объясните результат.


```python
def diceToss(number, k = 6):
    sumList,recordList= [], []
    rangeList = range(1, number+1)

    for i in rangeList:
        flip1 = random.randint(1,6) #возвращает число от 1 до 6
        flip2 = random.randint(1,6)
        flipsum = flip1 + flip2
        sumList.append(flipsum)
        recordList.append([flip1, flip2])
    res = sum([x > k for x in sumList])/number

    #print(f'Вероятность, что сумма на кубиках превысит {k} при {number} испытаниях: {res}'  )
    return res
```

Если рассмотреть географическую карту и случайным образом равномерно разбросать на ней точки с произвольными координатами, то окажется, что количество точек в определённом регионе будет пропорционально площади этого региона. Ниже представлен код, который делает следующее. На отрезке от −1 до 1 берутся два значения, `x` и `y`, из которых строится точка на плоскости с координатами (x, y). Если такая точка попадает внутрь единичного круга с центром в нуле, то такая точка отмечается синим цветом, в противном случае — красным. Количество точек задаётся параметром `n`.

Вызовите функцию для разных `n`. К чему будет стремиться доля синих точек? Как это связано с площадями геометрических фигур? Проанализируйте, как меняется доля синих точек с ростом количества точек, и сравните её с теоретическим значением площади.


```python
def plot_circle(n):
  x = 2 * np.random.random(int(n)) - 1
  y = 2 * np.random.random(int(n)) - 1
  insideX, insideY  = x[(x*x+y*y)<=1],y[(x*x+y*y)<=1] #координаты точек внутри
  outsideX, outsideY = x[(x*x+y*y)>1],y[(x*x+y*y)>1] #координаты точек снаружи

  print('Число точек внутри круга:', len(insideX))
  print('Число точек снаружи круга:', (len(outsideX)))
  print('Доля точек внутри круга:', len(insideX)/(len(outsideX)+len(insideX)))
  # добавил эту строчку для вычисления числа пи
  print('Число pi приближенно равно =', len(insideX)/(len(outsideX)+len(insideX))*4)

  #Построение графика
  fig, ax = plt.subplots(1)
  ax.scatter(insideX, insideY, c='b', alpha=0.6, edgecolor=None)
  ax.scatter(outsideX, outsideY, c='r', alpha=0.7, edgecolor=None)
  ax.set_aspect('equal')
  fig.show()
```

Доля синих точек с ростом их числа стремиться к четверти числа пи. Соответственно доля площади круга, вписанной в квадрат, к площади квадрата пропорциональна четверти пи.

В описание функции добавил строчку, вычисляющую приближенно число пи.


```python
# площадь круга единичного радиуса pi*R^2
np.pi
# площадь квадрата со стороной 2 =
4
# доля площади круга в площади квадрата со стороной 2 =
np.pi/4
# Мы можем приближенно посчитать значение числа pi =
# pi_ = 4* долю круга в квадрате
```

```python
def pi_approx(n):
  '''
  Функция, приближенно вычисляющая число pi
  '''
  x = 2 * np.random.random(int(n)) - 1
  y = 2 * np.random.random(int(n)) - 1
  insideX, insideY  = x[(x*x+y*y)<=1],y[(x*x+y*y)<=1] #координаты точек внутри
  outsideX, outsideY = x[(x*x+y*y)>1],y[(x*x+y*y)>1] #координаты точек снаружи

  return  len(insideX)/(len(outsideX)+len(insideX))*4
```

```python
# Колебания приближенных значений pi вокруг теоретического значения
n = 10_000
X = np.arange(10,n)
vec_pi_approx = np.vectorize(pi_approx)
plt.plot(vec_pi_approx(X))
plt.plot([np.pi]*n)
plt.title(f'Колебания приближенных значений pi вокруг теоретического значения при {n} точках')
plt.show()
```

```python
# приближенное значение pi
vec_pi_approx(X).mean()
```

```python
# распределение приближенных значений pi
plt.xlim(xmin = 3., xmax = 3.25)
plt.hist(vec_pi_approx(X),bins=100)
plt.title(f'Распределение приближенных значений pi при {n} точках')
plt.show()
```

Брошены два кубика. Какова вероятность, что произведение значений на кубиках превысит 20?

Вероятность того, что произведение двух случайных чисел будет больше 20, можно найти, используя формулу для произведения вероятности:

P(A ∩ B) = P(A) × P(B)

где A и B - события, которые могут произойти при бросании кубиков.

Для кубиков с гранями от 1 до 6, возможные значения будут 1, 2, 3, 4, 5 и 6. Вероятность выпадения каждого из этих значений равна 1/6.

Теперь рассмотрим событие A: произведение значений на двух кубиках больше 20.
Это может произойти только в том случае, если одно из значений на одном кубике больше ил равно 5, а другое значение на другом кубике больше или равно 4.

Таким образом, вероятность события A равна:

```python
import itertools

# словарь для подсчета произведения различных результатов  выпадения 6-и костей
# стороны костей помечены числами от 2 до 7, поскольку вклад единицы невелируется произведением
numb_bin_res_2 = list(itertools.product(np.arange(1,7), np.arange(1,7))) # декартово произведение 2 костей
dct = dict()
for elem in numb_bin_res_2:
    prod = np.prod(elem)
    if prod not in dct.keys():
        dct[prod] = 1
    else:
        dct[prod] += 1
dct = dict(sorted(dct.items())) # сортировка словаря по ключам
dct
```

```python
sum(dct.values()) # количество всех исходов
```

```python
import matplotlib.pyplot as plt

plt.plot(dct.keys(),dct.values())
plt.title('Частоты произведений результатов выпадения 6-и костей')
plt.show();
```

```python
import pandas as pd
list_res_prod = np.array(tuple(dct.items())) # словарь превратили в двумерный numpy-массив
list_res_prod = pd.DataFrame(list_res_prod, columns = ['Sum','Quontaty']) # сконвертили в датафрейм
res_df = list_res_prod[list_res_prod['Sum'] >20] # отобрали строки по условию
res_df['Quontaty'].sum()/sum(dct.values()) # нашли вероятность произведения > 20
```
