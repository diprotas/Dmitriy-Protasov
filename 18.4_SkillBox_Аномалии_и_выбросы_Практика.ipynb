```python
# Ваш код здесь
def calculate_outliers_3sigma (data):
    qntl_25 = data.quantile(.25)
    qntl_75 = data.quantile(.75)
    boundaries = (qntl_25 - 1.5*(qntl_75 - qntl_25), qntl_75 + 1.5*(qntl_75 - qntl_25))
    return boundaries
```
