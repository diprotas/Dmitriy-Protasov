## Задание 2

Для всех строк исходного датасета, сгруппированных по году выпуска автомобиля, посчитать среднее значение количества пройденных миль по группе. Результат необходимо добавить в новый столбец датафрейма.

```python
df_years_mean_distance = (
    df
    .groupby('year')['odometer']# группируем датафрейм по состоянию авто
    .agg(mean_distance='mean') #в полученных группах считаем среднее по количеству пройденных миль (odometr) и записываем в новый столбец mean_distance

    .reset_index()
)
df_years_mean_distance
```

```python
# Ваш код здесь
plt.figure(figsize=(4,2))
plt.plot(df_years_mean_distance.year,df_years_mean_distance.mean_distance, '.',color='coral')
plt.xlabel('Год')
plt.ylabel('Средний пробег')
plt.title('Средний пробег по годам выпуска');
```

## Ядерная регрессия, чтобы сгладить данные

```python
# данные для сглаживания
x = df_years_mean_distance.year
y = df_years_mean_distance.mean_distance
```

```python
# оптимальная ширина окна
b_opt = 1.059*x.std()*len(x)**(-1/5)
b_opt
```

```python
import numpy as np

# ядерные функции
def gaussian_kernel(z):
    '''функция Гаусса'''
    return (1/np.sqrt(2*np.pi))*np.exp(-0.5*z**2)

'''Вычислим веса и рассчитаем регрессию'''
def regressor_gauss(X, b, x = x, y = y): # здесь X - новое входное значение, для которого требуется рассчитать прогноз
    kernels = [gaussian_kernel((xi-X)/b) for xi in x] #вычисляем значение гауссова ядра для разности X и всех заданных точек
    weights = np.array([len(x) * (kernel/np.sum(kernels)) for kernel in kernels]) #вычисляем веса
    return np.dot(weights.reshape(len(weights)), y)/len(x)
```

```python
#нарисуем графики гауссовых ядерных регрессий для различных b
plt.figure(figsize=(8,5))
plt.plot(x, y,'.', alpha = 0.5, label = 'данные')

plt.plot(x, np.vectorize(regressor_gauss)(x, 4.5), label = '4.5')
plt.plot(x, np.vectorize(regressor_gauss)(x, 2), label = '2')
plt.plot(x, np.vectorize(regressor_gauss)(x, 1), label = '1')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Ядерная регрессия')
plt.legend()
plt.show()
```

```python
df_years_mean_distance['regressed_distance'] = np.vectorize(regressor_gauss)(x, 1)
```

```python
# Ваш код здесь
plt.figure(figsize=(8,5))
plt.plot(df_years_mean_distance.year,df_years_mean_distance.mean_distance, '.',color='skyblue')
plt.plot(df_years_mean_distance.year,df_years_mean_distance.regressed_distance, color='orange')
plt.xlabel('Год')
plt.ylabel('Средний пробег')
plt.title('Средний пробег по годам выпуска');
```
