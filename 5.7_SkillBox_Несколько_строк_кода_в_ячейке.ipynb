# Несколько строк кода в ячейке
Когда мы только начинали пользоваться Python, каждое новое вычисление писали в новой ячейке.

Потом мы стали иногда писать по 2-3 строчки кода в ячейке и запускать её целиком. Чем сложнее будут становиться наши программы, тем удобнее нам будет писать помногу строчек в одной ячейке. Некоторые куски кода будут неделимы на отдельные ячейки, другие мы будем компоновать по ячейкам из удобства.

Можно сравнить ячейку кода в ноутбуке с абзацем в тексте. Несколько предложений, объединённых общей мыслью (смыслом) или сюжетом. В программировании — несколько строк кода, некоторая последовательность действий, выражающая одну мысль, составляющая одну функцию.

Со временем вы сами поймёте, на какие куски удобнее делить код.

Запишем следующую задачу целиком в одной ячейке.

Предположим, вы собрались в путешествие на машине из Петербурга в Иркутск.

Навигатор говорит, что расстояние между городами — 5900 километров. Не всегда дороги на вашем пути будут хорошими, часто вы будете проезжать через населённые пункты — предположим, что ваша средняя скорость за весь путь будет 68 километров в час. Во время путешествия вы не хотите перенапрягаться, хотите останавливаться в красивых местах погулять, поэтому в день будете проезжать 4,5 часа. Вы знаете, что ваша машина потребляет по 8,3 литра бензина на 100 километров. Средняя цена на бензин — 45 рублей 80 копеек за литр.

Напишите код, который будет рассчитывать и выводить на экран количество полных дней путешествия (последний частичный день можно не считать), количество потраченного бензина, стоимость всего потраченного бензина.


```python
# Сохраняем всю необходимую информацию из условия в переменные
total_distance = 5900
average_velocity = 68
duration_per_day = 4.5
fuel_consumption = 8.3
average_cost_per_liter = 45.8


# Проводим необходимые вычисления
trip_duration = int(total_distance/(average_velocity * duration_per_day) )
fuel_consumption_per_day = average_velocity * duration_per_day * average_cost_per_liter

# Выводим результаты на экран
print("Путешествие будет длиться", trip_duration , "полных дней")
print("Количество потраченного бензина",total_distance * fuel_consumption,"литров")
print("Cтоимость всего потраченного бензина",total_distance * average_cost_per_liter,"рублей")
```
