```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/drive/MyDrive/SkillBox/Feature_Processing/vehicles_dataset_upd2.csv')
df
```

```python
df.odometer.describe()
```

```python
#форматирование вывода чисел
df.odometer.describe().apply(lambda x: f"{x:0.3f}")
```

```python
df.odometer
```

```python
df.odometer ** 2
```

```python
df.odometer ** 2 if df.odometer % 2 == 0 else 0
```

```python
df.apply(lambda x: x.odometer, axis=1)
```

```python
df.apply(lambda x: x.odometer ** 2, axis=1)
```

```python
# возведение в сетепень только четных чисел
df.apply(lambda x: x.odometer ** 2 if x.odometer % 2 == 0 else 0, axis=1)
```

```python
# перевод милей в метрическую систему
df['odometer_km'] = df.apply(lambda x: x.odometer  * 1.609344, axis=1)
```

```python
df['odometer_km']
```

```python
# корреляция
df['odometer_km'].corr(df['odometer'])
```

```python
# операции со значениями в строках
df['odometer/price'] = df.apply(lambda x: x.odometer / x.price, axis=1)
df['odometer/price']
```

```python
df['odometer/price'].corr(df['odometer'])
```

```python
df['odometer/price'].corr(df['price'])
```

```python
df['odometer'].corr(df['price'])
```
