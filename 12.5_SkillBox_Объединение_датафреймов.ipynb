```python
df11 = pd.DataFrame({'A': ['5', '87', '42', '43'],
                     'B': ['B0', 'B1', 'B2', 'B3'],
                     'C': ['C0', 'C1', 'C2', 'C3'],
                     'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
df12 = pd.DataFrame({'A1': ['43', '3', '80', '5'],
                     'B1': ['B0', 'B1', 'B5', 'B3'],
                     'C1': ['C4', 'C5', 'C6', 'C7'],
                     'D1': ['D4', 'D5', 'D6', 'D7']})
```

```python
pd.concat([df11,df12],axis = 1)
```

```python
pd.concat([df11,df12])
```

```python
df11.join(df12)
```

```python
# пересечение таблиц по равным элементам в столбцах B и B1
df11.merge(df12, left_on='B', right_on='B1', how='inner')
```

```python
pd.merge(left = df11, right = df12, left_on='B', right_on='B1', how='inner')
```

```python
# объединение таблиц по элементам в столбцах B и B1
df11.merge(df12, left_on='B', right_on='B1', how='outer')
```

```python
pd.merge(left = df11, right = df12, left_on='B', right_on='B1', how='outer')
```
