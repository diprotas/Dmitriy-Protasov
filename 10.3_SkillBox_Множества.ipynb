Коллекция (collection) — структура данных в Python, набор элементов.

Последовательность (sequence) — коллекция, элементы которой упорядочены.
К элементам последовательности можно обращаться по целочисленным индексам.

К последовательностям относят списки, строки и кортежи.

Словари и множества — неупорядоченные коллекции.

# Множества

Представим, что нам нужно посчитать количество уникальных пользователей сервиса за день. Какая структура данных подойдёт для этой задачи?

Можно положить всех залогиненных за сегодня пользователей в список. Но в списке могут быть дубли, в случае если кто-то из пользователей проходил процедуру авторизации несколько раз. Тогда придётся удалять дубли из списка и только потом считать количество. Это дополнительно потраченное время. Было бы здорово сразу записывать данные в такую структуру данных, которая, если видит, что такое значение уже существует, не записывает его повторно. Так мы сильно облегчим себе жизнь и сократим время на выполнение данной задачи.

Такая структура данных есть, это ***множество (set)***. Ключевая особенность множества заключается в том, что все его элементы уникальны. В отличии от кортежей и списков, в множестве элементы не упорядочены.

Множество можно сравнить с мешком, где содержимое расположено беспорядочно. К элементам множества нельзя обращаться по индексу, у них нет нумерации.


Чтобы создать пустое множество, нужно написать команду `set` и круглые скобки.

```python
test_set = set()
print(test_set)
print(type(test_set))
```

Других вариантов создать пустое множество нет.

Положить в множество значения можно двумя способами.

1. Внутрь объявления множества передать некий итерируемый объект: список, строку или кортеж. Попробуем передать список.

```python
test_set = set([1, 5, 10])
```

```python
print(test_set)
print(type(test_set))
```

Элементы в сете находятся в фигурных скобках. В списках были квадратные, в таплах круглые, а тут фигурные.

Теперь попробуем передать строку. Пусть это будет строка “mama”, где все буквы повторяются по два раза.

```python
test_set = set('mama')
```

```python
print(test_set)
print(type(test_set))
```

В сете оказались только уникальные элементы! Это очень удобно, когда вам нужно получить все уникальные элементы из списка или другого итерируемого объекта. Просто явно приводим его к множеству и можем просмотреть все уникальные элементы.

Напоследок попробуем передать в сет кортеж и убедиться, что он тоже превращается в множество.




```python
test_set = set((1, 2, 1))
print(test_set)
print(type(test_set))
```

Второй способ положить значения в сет – это перечислить их в фигурных скобках. Обратите внимание, что через фигурные скобки нельзя создать пустой сет. Но давайте попробуем.

```python
test_set = {}
print(test_set)
print(type(test_set))
```

С помощью команды `in` можно проверить, входит ли элемент в множество, так же как на списках и кортежах.

С помощью ключевого слова in можно выяснить, содержится ли целочисленное значение 42 в списке целочисленных значений, или проверить принадлежность строкового значения "21" множеству строковых значений.
Если x встречается в коллекции y, то мы будем говорить, что x — элемент
коллекции y.
Проверка принадлежности элемента множеству выполняется быстрее,
чем проверка принадлежности элемента списку: чтобы проверить наличие
элемента x в списке y, необходимо обходить весь список до тех пор, пока не
будет найден элемент x или не будут проверены все элементы. Множества же
реализованы аналогично словарям (или ассоциативным массивам): для проверки наличия элемента x в множестве y Python «под капотом» выполняет одну операцию
y[hash(x)] и проверяет, не равно ли None возвращаемое значение.

Разберём на примере, в чём же самая интересная часть работы со множествами.

Представьте, что вы работаете в онлайн-магазине техники. Клиенты магазина жалуются, что у вашего магазина недостаточно разнообразный ассортимент. Вам нужно проанализировать, какие позиции нужно добавить в каталог.

Первый этап вашего исследования достаточно простой: сравнить список категорий товаров нашего магазина (назовем его магазин А) со списком магазина-конкурента (назовем его магазин B) и найти:

* Процент категорий, совпадающих в двух магазинах, относительно всех категорий
* Процент категорий, который наш магазин (магазин А) не покрывает, относительно всех категорий
* Вывести категории, которые наш магазин (магазин А) не покрывает

В качестве входных данных поступает список категорий товаров магазина, в котором вы работаете, и список категорий магазина Б, который считали с сайта конкурента ваши коллеги-разработчики.



```python
store_A = ['телевизоры', 'планшеты', 'ноутбуки', 'смартфоны', 'компьютеры', 'аудиомагнитола']
store_B = ['ноутбуки', 'смартфоны', 'колонки', 'аксессуары', 'бытовая техника', 'мониторы', 'умный дом', 'ноутбуки']
```

Посмотрим на данные. Во время выгрузки могла произойти ошибка, и какие-то категории могли задвоиться. Например, мы видим, что во втором списке два раза встречается категория `ноутбуки`.

Чтобы провести наш анализ, нам нужно избавиться от дубликатов. Это можно сделать, преобразовав списки в множества.

```python
store_A_set = set(store_A)
store_B_set = set(store_B)
```

Выведем, какова была длина списков до преобразования в множества и какая стала после преобразования:

```python
print(f'ДО: {len(store_A)}, ПОСЛЕ: {len(store_A_set)}')
print(f'ДО: {len(store_B)}, ПОСЛЕ: {len(store_B_set)}')
```

Видим, что в нашем магазине 6 уникальных категорий, а в магазине B 7 уникальных категорий.

Первая задача – найти, какой процент категорий совпадает в обоих магазинах из всех доступных категорий. Чтобы посчитать это значение, нужно найти количество категорий, которые совпали в обоих множествах, и поделить его на общее количество категорий.  



```python
# пересечение двух множеств
```

Найдём сначала пересечение двух множеств: `store_A_set` и `store_B_set`. Для этого есть специальный метод пересечение. Чтобы его вызвать, нужно написать имя одного множества, потом через точку применить метод `.intersection()` и в круглых скобках указать имя второго множества.

```python
store_A_set.intersection(store_B_set)
```

Аналогичную операцию пересечения можно совершить с помощью знака & (И):

```python
store_A_set & store_B_set
```

Итого у нас две совпавшие категории: ноутбуки и смартфоны. Запишем результат в отдельную переменную `stores_intersection`.

```python
stores_intersection = store_A_set & store_B_set
```

Следующим шагом найдём, сколько всего категорий в двух магазинах. Для этого нужно склеить два множества или, говоря математическим языком, объединить их.

```python
# объединение двух множеств
```

Есть два абсолютно идентичных способа это сделать с помощью знака | (ИЛИ), либо с помощью метода `.union()`.




```python
store_A_set.union(store_B_set)
```

```python
store_A_set | store_B_set
```

```python
stores_union = store_A_set | store_B_set
```

Остаётся количество категорий, одинаковых у обоих магазинов, то есть в пересечении, поделить на количество элементов в объединении и умножить на 100.

```python
intersection_percent = (len(stores_intersection)/len(stores_union))*100
print(f'Процент совпавших категорий: {round(intersection_percent,1)} %')
```

Следующая задача – найти процент категорий, который не покрыт нашим магазином, относительно всех категорий. Эту задачку можно было бы решить математически: знаем, что в магазине B 7 категорий, 2 категории совпали, значит — не совпали всего 5. И это значение нужно поделить на общее количество категорий.

Но магазинов, которые нужно сравнить, может быть не два, а три или ещё больше. И проводить вычисления в уме или на бумаге станет всё сложнее. Поэтому давайте отработаем решение с помощью методов работы с множествами.


```python
# вычитание одного множества из другого
```

Найдём категории, которые есть у магазина B, но нет у магазина А. Эта операция называется **difference** или вычитание. Для неё, конечно же, тоже есть специальный метод. На первом месте пишем то множество, из которого хотим вычесть элементы. В нашем случае это магазин B.  И через минус — второе множество.



```python
store_B_set - store_A_set
```

Второй способ — с помощью метода `.difference()`. На первом месте опять же магазин B, в скобках — магазин A.

```python
store_B_set.difference(store_A_set)
```

Итак, у нашего магазина отсутствуют аксессуары, бытовая техника, колонки, мониторы и всё, что относится к умному дому.

Аналогично предыдущему пункту сохраним значения в переменную `B_dif_A`.

Далее посчитаем процент непокрытых нашим магазином товаров.

```python
B_dif_A = store_B_set - store_A_set
B_dif_A_percent = (len(B_dif_A)/len(stores_union))*100
print(f'Процент непокрытых категорий из B: {round(B_dif_A_percent,1)} %')
```

Получили, что из всех возможных категорий у нашего магазина 45 % не покрыто.

18 % категорий есть и у нас, и у магазина B.

100 - 18 - 45 = 37

37 % категорий есть у нас, но нет у магазина конкурента. Давайте проверим, посчитав вычитание между магазином А и магазином B.

```python
A_dif_B = store_A_set - store_B_set
A_dif_B_percent = (len(A_dif_B)/len(stores_union))*100
print(f'Процент непокрытых категорий из A: {round(A_dif_B_percent,2)} %')
```

Есть ещё один метод над множествами, который мы не рассмотрели в текущей задаче. Он называется симметричное вычитание или symmetric_difference. Он ищет все элементы в A и B, которые не входят в пересечение двух множеств. Давайте попробуем его применить на сетах магазинов. Есть два способа нахождения симметричного вычитания: с помощью метода `.symmetric difference()` и с помощью знака ^.



```python
store_B_set.symmetric_difference(store_A_set)
```

```python
store_B_set ^ store_A_set
```

Так как вычитание симметричное, если мы изменим порядок аргументов — результат не изменится.

Множество — коллекция элементов подобно спискам и кортежам. Коллекция состоит либо из элементов простых типов данных (целочисленных
значений, значений с плавающей точкой, строковых значений), либо из
более сложных элементов (объектов, кортежей). Однако все типы данных
в множестве должны быть хешируемыми, то есть обладать соответствующим
хеш-значением.
Хеш-значение объекта никогда не меняется и используется для его сравнения с другими объектами. Рассмотрим пример в листинге 1.8, где множество создается на основе трех строковых значений, после проверки их
хеш-значений. Далее пробуем создать множество списков, но это нам не
удается, поскольку списки нехешируемые.
Листинг 1.8. Множество требует хешируемых элементов

```python
hero = "Harry"
guide = "Dumbledore"
enemy = "Lord V."
print(hash(hero))
# 6175908009919104006
print(hash(guide))
# -5197671124693729851
## Можно ли создать множество строковых значений?
characters = {hero, guide, enemy}
print(characters)
# {'Lord V.', 'Dumbledore', 'Harry'}
```

Множество строковых значений можно создать, поскольку строковые значения — хешируемые. А создать множество списков нельзя, поскольку списки нехешируемые. Дело в том, что хеш-значение зависит от содержимого элемента коллекции, а списки — изменяемые; если модифицировать данные в списке, то хеширование тоже должно измениться. А поскольку изменяемые типы данных нехешируемы, использовать их в множествах нельзя.
