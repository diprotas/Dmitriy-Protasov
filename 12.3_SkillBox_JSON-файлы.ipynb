Чтение json в датафрейм pandas:

```python
help(pd.read_json)
```

```python
pd.read_json('12.3.Data/partner_data_records.json')
```

Аргумента `index_col` у функции нет:

```python
pd.read_json('12.3.Data/partner_data_records.json', index_col='id')
```

С помощью специального метода `set_index` укажем, какую колонку использовать в качестве индекса.

```python
df = pd.read_json('12.3.Data/partner_data_records.json')
df
```

```python
df.set_index('id')
```

Однако изменения в датафрейме не сохраняются:

```python
df.head()
```

Аргумент `inplace=True` позволяет сохранить изменения:

```python
df = pd.read_json('12.3.Data/partner_data_records.json')
df.set_index('id', inplace=True)
```

```python
df.head()
```

Можно также сразу переписывать объект нашего датафрейма через `df=`:


```python
df = pd.read_json('12.3.Data/partner_data_records.json')
df = df.set_index('id')
df.tail()
```
