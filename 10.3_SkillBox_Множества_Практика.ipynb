*Задача 1. Операции над множествами*

Есть два списка. В списке `opened` лежат все ID договоров, открытых за прошлую неделю. В списке `closed` лежат все ID договоров, закрытых за прошлую неделю.

Найдите с помощью множеств следующие значения:
1. ID договоров, которые были открыты, а затем закрыты на прошлой неделе.
2. ID договоров, которые были закрыты на прошлой неделе, но открыты ранее прошлой недели.
3. ID договоров, которые были открыты на прошлой неделе, но пока не закрылись.
4. Все ID договоров, с которыми производили операции закрытия или открытия на прошлой неделе.
5. Проверьте, что сумма элементов из 1, 2, 3 равна 4.

Проверьте работоспособность программы на разных значениях `opened`, `closed`.






```python
opened = ['123', '156', '189', '201', '222', '256', '266', '287', '289', '300']
closed = ['20', '78', '85', '123', '222', '256', '287', '289']
```

```python
opened.sort(key = lambda x: x in closed)
# сначала отсортированы элементы отсутствующие в closed: '156', '189', '201', '266', '300'
# а затем идут отсортированные элементы присутствующие в closed: '123', '222', '256', '287', '289'
opened # ['156', '189', '201', '266', '300', '123', '222', '256', '287', '289']
```

```python
# Ваш код здесь
set_opened = set(opened)
set_closed = set(closed)
print(f'         ID договоров, которые были открыты, а затем закрыты на прошлой неделе:{set_opened & set_closed}.\n \
        ID договоров, которые были закрыты на прошлой неделе, но открыты ранее прошлой недели:{set_closed - set_opened }.\n \
        ID договоров, которые были открыты на прошлой неделе, но пока не закрылись:{set_opened - set_closed}.\n \
        Все ID договоров, с которыми производили операции закрытия или открытия на прошлой неделе:{set_opened ^ set_closed}.\n \
        Проверьте, что сумма элементов из 1, 2, 3 равна 4:{len(set_opened & set_closed)+len(set_closed - (set_opened & set_closed))+len(set_opened - set_closed)}.')
```

*Задача 2. Проверка на дубли*

Дан список `numbers`. Для каждого элемента выведите, встречался он до этого в списке или нет. Используйте множества для решения этой задачи.

Проверьте работоспособность программы на разных значениях `numbers`.


Пример:
<br>
Input: `numbers = [1, 2, 1, 3, 2]`
<br>
Output:

`1 - не встречалось`

`2 - не встречалось`

`1 - дубль`

`3 - не встречалось`

`2 - дубль`

```python
numbers = [1, 2, 1, 3, 2]
```

```python
# Ваш код здесь
def has_dubles(numbers):
  dct =dict()
  for elem in numbers:
    if elem in dct.keys():
      dct[elem] += 1
      print(f'{elem} - дубль')
    else:
      dct[elem] = 1
      print(f'{elem} - не встречалось')
  return dct
```

```python
has_dubles(numbers)
```

```python
for i in numbers:
  print(f'{i}:{numbers.count(i)}')
```

*Задача 3. Проверка на плагиат*

Даны два текста, которые лежат в переменных `text1` и  `text2` соответственно. Найдите схожесть данных текстов. Давайте считать схожестью процент, который можно найти по следующей формуле:

схожесть = (количество уникальных одинаковых слов / общее число уникальных слов в двух текстах) * 100%


В этом задании слова не нужно приводить в начальную форму. Так, "цены" и "цен" в текстах ниже мы будем считать разными словами.

Подсказка:
1. От нежелательных знаков препинания можно избавиться с помощью `replace`.
2. Символ `\` удалять не нужно, он позволяет описывать строковую переменную в несколько строк в коде.


```python
text1 = 'В первом квартале года отмечался рост цен практически на все сырьевые товары, что отражало сохранение тенденции к заметному росту цен, начавшемуся в середине 2020 года.\
 Больше всего подорожали энергоресурсы, что имеет особое значение для российского экспорта топливо-энергетических товаров.'

text2 = 'Цены на неэнергетические сырьевые товары в третьем квартале года в среднем выросли на 10% и по итогам года они покажут незначительный рост по сравнению с 2020. \
Это приведет к заметному росту цен на другие ресурсы.'
```

```python
norm_text1 = text1.lower().replace('.','').replace(',','').split()
norm_text2 = text2.lower().replace('.','').replace(',','').split()
set_norm_text1 = set(norm_text1)
set_norm_text2 = set(norm_text2)
union_set = set_norm_text1 | set_norm_text2
union_set
```

```python
# Ваш код здесь
from itertools import chain

def plagiat(text1,text2):
  '''
  схожесть = (количество уникальных одинаковых слов / общее число уникальных слов в двух текстах) * 100%

  переработанная формула схожести
  схожесть = (количество уникальных одинаковых слов - количество полных пар словоформ в пересечении)/(общее число уникальных слов в двух текстах - количество полных пар словоформ) * 100%
  '''
  norm_text1 = text1.lower().replace('.','').replace(',','').split()
  norm_text2 = text2.lower().replace('.','').replace(',','').split()
  set_norm_text1 = set(norm_text1)
  set_norm_text2 = set(norm_text2)
  intersec_set = set_norm_text1 & set_norm_text2
  intersec_set = {word for word in intersec_set if len(word) > 2} # удалил стоп-слова длины <= 2
  union_set = set_norm_text1 | set_norm_text2 # объединение множеств
  print('пересечение: ',intersec_set)

  # поиск слов длиною >2, входящих в другие слова, отличающихся от них на один символ
  set_any_morphem =set([(word1,word) for word in union_set for word1 in union_set if (word != word1)\
     and (word.find(word1) != -1) and len(word1)>2 and abs(len(word) -len(word1))<2])
  print('общее количество пар словоформ (в объединении): ',len(set_any_morphem))
  print(set_any_morphem)
  set_any_morphem = set(chain(*set_any_morphem)) # выпрямляю множество морфем

  print('количество пар словоформ в пересечении: ',len(intersec_set & set_any_morphem)//2)
  return f'{round(100*(len(intersec_set) - len(intersec_set & set_any_morphem)//2)/(len(union_set) - len(set_any_morphem)),2)}%'
```

```python
# поиск слов, входящих в другие слова
set([word for word in union_set for word1 in union_set \
     if (word1 != word) and (word1.find(word) != -1) and len(word)>2])
```

```python
# поиск морфем (словоформ) в тексте
# поиск слов, в которые вошли другие слова, но разность длин которых < 2



set_any_morphem =set([(word1,word) for word in union_set for word1 in union_set if (word != word1)\
     and (word.find(word1) != -1) and len(word1)>2 and abs(len(word) -len(word1))<2])

set(chain(*set_any_morphem))
```

```python
numbers = [1, 2, 1, 3, 2]
```

```python
set_numbers = set()
for el in numbers:
    if el in set_numbers:
        print(el, '- дубль')
    else:
        print(el, '- не встречалось')
        set_numbers.add(el)
```
