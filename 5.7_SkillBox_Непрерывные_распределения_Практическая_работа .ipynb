## –ó–∞–¥–∞—á–∏ –ø–æ —Ç–µ–º–µ ¬´–û—Å–Ω–æ–≤–Ω—ã–µ –≤–∏–¥—ã –Ω–µ–ø—Ä–µ—Ä—ã–≤–Ω—ã—Ö —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–π¬ª
### 1.1. –ó–∞–¥–∞—á–∞

–î–∞–Ω–∞ —Ñ—É–Ω–∫—Ü–∏—è —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –Ω–µ–ø—Ä–µ—Ä—ã–≤–Ω–æ–π –≤–µ–ª–∏—á–∏–Ω—ã:

$$
F(x) = \left\{
    \begin{array}\\
        0 & \mbox{–ø—Ä–∏ } & x < 1, \\
        \frac{1}{4}(x - 1) & \mbox{–ø—Ä–∏ } & c \leq x \leq d, \\
        1 & \mbox{–ø—Ä–∏ } & x > 5.
    \end{array}
\right.
$$

–ù–µ–æ–±—Ö–æ–¥–∏–º–æ –Ω–∞–π—Ç–∏:

1. –∑–Ω–∞—á–µ–Ω–∏—è $c$ –∏ $d$;
2. –ø–ª–æ—Ç–Ω–æ—Å—Ç—å —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–µ–π $f(x)$;
3. –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ–∂–∏–¥–∞–Ω–∏–µ $M(x)$;
4. –¥–∏—Å–ø–µ—Ä—Å–∏—é $D(x)$;
5. –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –ø–æ–ø–∞–¥–∞–Ω–∏—è —Å–ª—É—á–∞–π–Ω–æ–π –≤–µ–ª–∏—á–∏–Ω—ã –≤ –æ—Ç—Ä–µ–∑–æ–∫ $[1;2]$.

–£–∫–∞–∑–∞–Ω–∏–µ: –≤—Å–ø–æ–º–Ω–∏—Ç–µ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –∏ —Å–≤–æ–π—Å—Ç–≤–∞ —Ñ—É–Ω–∫—Ü–∏–π —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –∏ –ø–ª–æ—Ç–Ω–æ—Å—Ç–∏ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–µ–π.

–ù–∞–π–¥–∏—Ç–µ:

1. –ø–ª–æ—Ç–Ω–æ—Å—Ç—å —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è $f(x)$;
2. –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ–∂–∏–¥–∞–Ω–∏–µ $M(X)$;
3. –¥–∏—Å–ø–µ—Ä—Å–∏—é $D(X)$ –∏ —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ $\sigma(X)$.

### –†–µ—à–µ–Ω–∏–µ

#####  1. –ó–Ω–∞—á–µ–Ω–∏—è –∫–æ–Ω—Å—Ç–∞–Ω—Ç
–∑–Ω–∞—á–µ–Ω–∏—è $c=1 $ –∏ $d=5$;

#####  2. –ü–ª–æ—Ç–Ω–æ—Å—Ç—å —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è

$$
f(x) = F'(x) = \left\{
    \begin{array}\\
        0 & \mbox{–ø—Ä–∏ } & x < 1, \\
        \frac{1}{4} & \mbox{–ø—Ä–∏ } & 1 \leq x \leq 5, \\
        0 & \mbox{–ø—Ä–∏ } & x > 5.
    \end{array}
\right.
$$

#####  3. –ú–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ–∂–∏–¥–∞–Ω–∏–µ

$$M(X) = \int_{-\infty}^{\infty} x f(x) dx = \int_1^5 x f(x) dx = \frac{1}{4} \int_1^5 x dx = \frac{1}{8} (\frac{5^2}{1} - \frac{1^2}{1}) = \frac{24}{8} = 3$$

#####  4. –î–∏—Å–ø–µ—Ä—Å–∏—è –∏ —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ

$$D(X) = \int_{-\infty}^{\infty} x^2 f(x) dx = \int_1^5 x^2 f(x) dx = \frac{1}{4} \int_1^5 x^2 dx = \frac{1}{4} (\frac{5^3}{3} - \frac{1^3}{3}) = \frac{124}{12} \approx 10.3.$$

$$\sigma(X) = \sqrt{D(X)} = \sqrt{10.3}   \approx 3.21.$$

#####  5. –í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –ø–æ–ø–∞–¥–∞–Ω–∏—è —Å–ª—É—á–∞–π–Ω–æ–π –≤–µ–ª–∏—á–∏–Ω—ã –≤ –æ—Ç—Ä–µ–∑–æ–∫ $[1;2]$

$$P(a Ôºú X Ôºú b) = \int_{a}^{b} f(x) dx = \frac{1}{4} \int_{1}^{2} dx = \frac{1}{4} $$

### 1.2. –ó–∞–¥–∞—á–∞

–ò–∑–≤–µ—Å—Ç–Ω–æ, —á—Ç–æ –≤—Å–µ –∑–Ω–∞—á–µ–Ω–∏—è —Ä–∞–≤–Ω–æ–º–µ—Ä–Ω–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª—ë–Ω–Ω–æ–π —Å–ª—É—á–∞–π–Ω–æ–π –≤–µ–ª–∏—á–∏–Ω—ã –ª–µ–∂–∞—Ç –≤ –æ—Ç—Ä–µ–∑–∫–µ $[3;9]$. –ù–∞–π–¥–∏—Ç–µ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –ø–æ–ø–∞–¥–∞–Ω–∏—è —ç—Ç–æ–π —Å–ª—É—á–∞–π–Ω–æ–π –≤–µ–ª–∏—á–∏–Ω—ã –≤ –ø—Ä–æ–º–µ–∂—É—Ç–æ–∫ $(5;7)$.

–£–∫–∞–∑–∞–Ω–∏–µ: –≤—Å–ø–æ–º–Ω–∏—Ç–µ, –∫–∞–∫–æ–π –≤–∏–¥ –∏–º–µ–µ—Ç —Ñ—É–Ω–∫—Ü–∏—è –ø–ª–æ—Ç–Ω–æ—Å—Ç–∏ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–µ–π —Ä–∞–≤–Ω–æ–º–µ—Ä–Ω–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª—ë–Ω–Ω–æ–π –≤–µ–ª–∏—á–∏–Ω—ã.



$$
f(x) =  \left\{
    \begin{array}\\
        0 & \mbox{–ø—Ä–∏ } & x < 3, \\
        \frac{1}{6} & \mbox{–ø—Ä–∏ } & 3 \leq x \leq 9, \\
        0 & \mbox{–ø—Ä–∏ } & x > 9.
    \end{array}
\right.
$$

–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –ø–æ–ø–∞–¥–∞–Ω–∏—è —Å–ª—É—á–∞–π–Ω–æ–π –≤–µ–ª–∏—á–∏–Ω—ã –≤ –ø—Ä–æ–º–µ–∂—É—Ç–æ–∫ $(5;7)$

$$P(Œ± Ôºú X Ôºú Œ≤) = \frac{Œ± ‚àí ùõΩ}{ùëé ‚àí ùëè} = \frac{7 ‚àí 5}{6} = \frac{1}{3}$$


### 1.3. –ó–∞–¥–∞—á–∞

–°–ª—É—á–∞–π–Ω–∞—è –≤–µ–ª–∏—á–∏–Ω–∞ $X$ –∏–º–µ–µ—Ç –Ω–æ—Ä–º–∞–ª—å–Ω–æ–µ  —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Å –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–∏–º –æ–∂–∏–¥–∞–Ω–∏–µ–º $a = 5$ –∏ –¥–∏—Å–ø–µ—Ä—Å–∏–µ–π $D(x) = 9$. –ù–∞–π–¥–∏—Ç–µ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –ø–æ–ø–∞–¥–∞–Ω–∏—è —ç—Ç–æ–π —Å–ª—É—á–∞–π–Ω–æ–π –≤–µ–ª–∏—á–∏–Ω—ã –≤ –∏–Ω—Ç–µ—Ä–≤–∞–ª $(7;12)$.

–£–∫–∞–∑–∞–Ω–∏–µ: –≤—Å–ø–æ–º–Ω–∏—Ç–µ, –∫–∞–∫ —Å–≤—è–∑–∞–Ω–æ –Ω–æ—Ä–º–∞–ª—å–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Å —Ñ—É–Ω–∫—Ü–∏–µ–π –õ–∞–ø–ª–∞—Å–∞.

–ü–ª–æ—Ç–Ω–æ—Å—Ç—å –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ –Ω–æ—Ä–º–∞–ª—å–Ω–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–Ω–æ–π —Å–ª—É—á–∞–π–Ω–æ–π –≤–µ–ª–∏—á–∏–Ω—ã $X$ –æ–ø–∏—Å—ã–≤–∞–µ—Ç—Å—è —Ñ—É–Ω–∫—Ü–∏–µ–π –ì–∞—É—Å—Å–∞:

$$f(x) = \frac{1}{\sigma \sqrt{2\pi}} \exp{(-\frac{(x-\mu)^2}{2\sigma^2})},$$

–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –ø–æ–ø–∞–¥–∞–Ω–∏—è –∑–Ω–∞—á–µ–Ω–∏—è —Å–ª—É—á–∞–π–Ω–æ–π –≤–µ–ª–∏—á–∏–Ω—ã $X$ –≤ –∏–Ω—Ç–µ—Ä–≤–∞–ª $(Œ±; Œ≤)$

$$P(\alpha < X < \beta) = \Phi(\frac{\beta-\mu}{\sigma}) - \Phi(\frac{\alpha-\mu}{\sigma}).$$

–í –Ω–∞—à–µ–º —Å–ª—É—á–∞–µ
$$P(7 < X < 12) = \Phi(\frac{12-5}{3}) - \Phi(\frac{7 -5}{3}) = \Phi(\frac{7}{3}) - \Phi(\frac{2}{3}) = \Phi(2.33) - \Phi(0.67) = 0.49 - 0.2486 = 0.2418$$

### 2.1. –ó–∞–¥–∞—á–∞

–ò—Å–ø–æ–ª—å–∑—É—è –º—É–ª—å—Ç–∏–ø–ª–∏–∫–∞—Ç–∏–≤–Ω—ã–π –º–µ—Ç–æ–¥ —Ñ–æ—Ä–º–∏—Ä–æ–≤–∞–Ω–∏—è —Å–ª—É—á–∞–π–Ω–æ–π –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç–∏, —Ä–∞—Å—Å–º–æ—Ç—Ä–µ–Ω–Ω—ã–π –≤ —à–µ—Å—Ç–æ–π —Ç–µ–º–µ –º–æ–¥—É–ª—è, –ø–æ—Å—Ç—Ä–æ–π—Ç–µ –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å —Å—Ç–∞ —Å–ª—É—á–∞–π–Ω—ã—Ö —á–∏—Å–µ–ª.

–£–∫–∞–∑–∞–Ω–∏–µ: –¥–ª—è —Ä–µ–∞–ª–∏–∑–∞—Ü–∏–∏ –∑–∞–¥–∞—á–∏ –∏—Å–ø–æ–ª—å–∑—É–π—Ç–µ —Ü–∏–∫–ª. –ü–µ—Ä–≤—ã–π —á–ª–µ–Ω –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç–∏ $X_0$ –∑–∞–¥–∞—ë—Ç—Å—è –ø—Ä–æ–∏–∑–≤–æ–ª—å–Ω—ã–º –æ–±—Ä–∞–∑–æ–º. –¢–∞–∫–∂–µ –æ–±—Ä–∞—Ç–∏—Ç–µ –≤–Ω–∏–º–∞–Ω–∏–µ, —á—Ç–æ —Ä–µ–∞–ª–∏–∑–æ–≤–∞—Ç—å –¥–∞–Ω–Ω—ã–π –∞–ª–≥–æ—Ä–∏—Ç–º –º–æ–∂–Ω–æ –Ω–∞ –ª—é–±–æ–º —É–¥–æ–±–Ω–æ–º —è–∑—ã–∫–µ –ø—Ä–æ–≥—Ä–∞–º–º–∏—Ä–æ–≤–∞–Ω–∏—è. –í —Å–ª—É—á–∞–µ –ø—Ä–∞–≤–∏–ª—å–Ω–æ–π —Ä–µ–∞–ª–∏–∑–∞—Ü–∏–∏ –∞–ª–≥–æ—Ä–∏—Ç–º–∞ –Ω–∞ –≤—ã—Ö–æ–¥–µ –≤—ã –ø–æ–ª—É—á–∏—Ç–µ —Å—Ç–æ —Å–ª—É—á–∞–π–Ω—ã—Ö —á–∏—Å–µ–ª. –ß–∏—Å–ª–∞ –≥–µ–Ω–µ—Ä–∏—Ä—É—é—Ç—Å—è —Å–ª—É—á–∞–π–Ω—ã–º –æ–±—Ä–∞–∑–æ–º, –∑–Ω–∞—á–∏—Ç, –ø—Ä–∏ –∫–∞–∂–¥–æ–º –Ω–æ–≤–æ–º –∑–∞–ø—É—Å–∫–µ –∞–ª–≥–æ—Ä–∏—Ç–º–∞ –≤—ã –±—É–¥–µ—Ç–µ –ø–æ–ª—É—á–∞—Ç—å –Ω–æ–≤—É—é –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å.



```python
import random

def multiplicat_generator(n):
    x = random.random() * 10**6
    res_list = [x]
    t = 1
    l_ambda = 8*t - 3
    p = 10
    q = 3
    m = p**q
    for i in range(n):
        x = l_ambda * (x % m)
        x /= m
        res_list.append(x)
    return res_list
```

```python
multiplicat_generator(100)
```

### 2.2. –ó–∞–¥–∞—á–∞

–ù–µ–∫–æ—Ç–æ—Ä–∞—è —Å–ª—É—á–∞–π–Ω–∞—è –≤–µ–ª–∏—á–∏–Ω–∞ –∑–∞–¥–∞–Ω–∞ —ç–∫—Å–ø–æ–Ω–µ–Ω—Ü–∏–∞–ª—å–Ω—ã–º –∑–∞–∫–æ–Ω–æ–º —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è:

$$
f(y) =  \left\{
    \begin{array}\\
        0, & \mbox{–ø—Ä–∏ } & y < 0, \\
        4 e^{-4y}, & \mbox{–ø—Ä–∏ } & y \geq 0,         
    \end{array}
\right.
$$

–°–æ—Å—Ç–∞–≤—å—Ç–µ –≥–µ–Ω–µ—Ä–∞—Ü–∏—é –¥–µ—Å—è—Ç–∏ —Å–ª—É—á–∞–π–Ω—ã—Ö —á–∏—Å–µ–ª $y_i$.

```python
import numpy as np

def exponent_generator(n):
    y = random.random()
    res_list =[y]
    t = 1
    l_ambda = 8*t - 3
    p = 10
    q = 1
    m = p**q
    for i in range(n):
        y = l_ambda * (y % m)
        y /= m
        y = -np.log10(1 - y) / 4
        res_list.append(y)
    return res_list
```

```python
exponent_generator(10)
```

```python
import math
# –î–∞–Ω–Ω—ã–π –∫–æ–¥ –∏—Å–ø–æ–ª—å–∑—É–µ—Ç –∫–æ–Ω–≥—Ä—É—ç–Ω—Ç–Ω—ã–π –º–µ—Ç–æ–¥ –¥–ª—è –≥–µ–Ω–µ—Ä–∞—Ü–∏–∏ –ø—Å–µ–≤–¥–æ—Å–ª—É—á–∞–π–Ω—ã—Ö —á–∏—Å–µ–ª, –∞ –∑–∞—Ç–µ–º –ø—Ä–∏–º–µ–Ω—è–µ—Ç —ç–∫—Å–ø–æ–Ω–µ–Ω—Ü–∏–∞–ª—å–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –∫ –ø–æ–ª—É—á–µ–Ω–Ω—ã–º —á–∏—Å–ª–∞–º.
# –ø—Ä–∏–º–µ—Ä –∫–æ–¥–∞ –Ω–∞ Python –¥–ª—è –≥–µ–Ω–µ—Ä–∞—Ü–∏–∏ –¥–µ—Å—è—Ç–∏ —Å–ª—É—á–∞–π–Ω—ã—Ö —á–∏—Å–µ–ª yi —Å–æ–≥–ª–∞—Å–Ω–æ —ç–∫—Å–ø–æ–Ω–µ–Ω—Ü–∏–∞–ª—å–Ω–æ–º—É –∑–∞–∫–æ–Ω—É —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è, –∏—Å–ø–æ–ª—å–∑—É—è –∫–æ–Ω–≥—Ä—É—ç–Ω—Ç–Ω—ã–π –º–µ—Ç–æ–¥:

# –ù–∞—á–∞–ª—å–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ X0
X0 = 12345
# –ö–æ–Ω—Å—Ç–∞–Ω—Ç—ã –¥–ª—è –∫–æ–Ω–≥—Ä—É—ç–Ω—Ç–Ω–æ–≥–æ –º–µ—Ç–æ–¥–∞
a = 22695477
c = 1
m = 2**32
n = 10  # –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –≥–µ–Ω–µ—Ä–∏—Ä—É–µ–º—ã—Ö —Å–ª—É—á–∞–π–Ω—ã—Ö —á–∏—Å–µ–ª

# –ì–µ–Ω–µ—Ä–∞—Ü–∏—è –ø—Å–µ–≤–¥–æ—Å–ª—É—á–∞–π–Ω—ã—Ö —á–∏—Å–µ–ª —Å –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ–º –∫–æ–Ω–≥—Ä—É—ç–Ω—Ç–Ω–æ–≥–æ –º–µ—Ç–æ–¥–∞
random_sequence = [X0]
for i in range(1, n):
    Xi = (a * random_sequence[i-1] + c) % m
    random_sequence.append(Xi)

# –ü—Ä–∏–º–µ–Ω–µ–Ω–∏–µ —Ñ—É–Ω–∫—Ü–∏–∏ —ç–∫—Å–ø–æ–Ω–µ–Ω—Ü–∏–∞–ª—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è
exponential_random_numbers = [(-1/4) * (1/4) * math.log(1-random_sequence[i]/m) if random_sequence[i] !=0  else 0 for i in range(n)]

# –í—ã–≤–æ–¥ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞
exponential_random_numbers
```
